```python
df.dropna(inplace=True)
value = 0
df.fillna(value, inplace=True)
```

```python
sns.set_style("darkgrid")
```

```python
plt.figure(figsize=(20 , 4))
sns.histplot(df['Date'], bins=20, kde=True, color='blue')
plt.title("Temperature")
plt.xlabel("Rainfall")
plt.ylabel("humidity")
plt.show()


sns.histplot(df['Temperature'], kde=True, bins=30, color='blue')
plt.title("Temperature Distribution")
plt.xlabel("Temperature (Â°C)")
plt.ylabel("Frequency")
plt.show()

sns.histplot(df['Humidity'], kde=True,bins=30, color='blue')
plt.title('Humidity')
plt.xlabel('Humidity')
plt.ylabel('Frequency')
plt.show()


sns.histplot(df['Rainfall'], kde=True,bins=30, color='blue')
plt.title('Rainfall')
plt.xlabel('Rainfall')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['Temperature'], bins=10)
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Temperature')
plt.show()

plt.scatter(df['Temperature'], df['Rainfall'])
plt.xlabel('Temperature')
plt.ylabel('Rainfall')
plt.title('Temperature vs Rainfall')
plt.show()

sns.boxplot(x=df['Temperature'], y=df['Humidity'])
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.title('Temperature vs Humidity')
plt.show()
```
